```python
def visualize(filename):
    "Plot the Features"
    data = pd.read_csv(filepath_or_buffer = filename)
    average = data.mean(axis=0)
    minimum = data.min(axis=0)
    maximum = data.max(axis=0)
    fig = plt.figure(figsize = (12,12))
    x = list(range(1,41))
    plt.plot(x, average, color = 'blue', label = 'average')
    plt.plot(x, minimum, color = 'green', label ='minimum')
    plt.plot(x, maximum, color ='red', label ='maximum')
    plt.xlabel('Feature#')
    plt.ylabel('Values')
    plt.show()
```

```python
def issues(filename):
    data = pd.read_csv(filepath_or_buffer = filename)
    minimum = data.min(axis=0)
    maximum = data.max(axis=0)
    if maximum[0]==0 and maximum[20]==20 :
        print('I doubt the integrity of this Dataset')
    else:
        if  minimum.sum(axis=0) == 0 :
            print('why is this happening')
        else :
            print('Dataset looks good')
```
